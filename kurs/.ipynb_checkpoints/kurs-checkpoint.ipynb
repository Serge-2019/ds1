## Подготовка данных

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
df.describe()
```

```python
# есть несколько записей со странными годами я приравнял их к 2020 кстати год 2020 
# я так понял предполагает что дом еще будет достроен к этому году

df.loc[df['HouseYear']>2019, 'HouseYear'] = 2020
df.loc[df['HouseFloor'] < 1, 'HouseFloor'] = 1

df
```

```python
df[df['Rooms'] < 1]
```

```python
r = df[(df['Square']>37) & (df['Square']<39)  & (df['Rooms']!=0)].copy()
r['y'] = r['HouseYear'] // 10

r.pivot_table(values='HouseYear', index='Rooms', columns='y', aggfunc='count')
```
