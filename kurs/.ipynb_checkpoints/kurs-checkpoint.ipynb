# Подготовка данных

**Вспомогательные таблицы**

````python
def fltLifeSquare(d):
    return (d['LifeSquare'].isnull()) \
            | (d['LifeSquare'] >= d['Square']) \
            | (d['LifeSquare'] < d['Square'] * 0.2)


def fltKitchenSquare(d):
    return (d['Square'] < d['KitchenSquare']) \
            | (d['KitchenSquare'] <= 1) \
            | (d['KitchenSquare'] > d['Square']*0.3)


def getMeans(d):
    i = d['HouseFloor'] < d['Floor']
    d.loc[i, 'HouseFloor'] = d.loc[i, 'Floor']
    r = d.groupby('HouseYear', as_index=False).agg({'HouseFloor': 'mean'}).rename(columns={'HouseFloor': 'mhf'})
    
    # LifeSquare
    ind = fltLifeSquare(d)
    f = d[~ind].groupby('HouseYear', as_index=False).agg({'LifeSquare': 'mean', 'Square': 'mean'})
    f['%lif'] = f['LifeSquare'] / f['Square']
    f.drop(['LifeSquare', 'Square'], axis=1, inplace=True)
    r = r.merge(f, on='HouseYear', how='left')
    
    # KitchenSquare
    ind = fltKitchenSquare(d)
    f = d[~ind].groupby('HouseYear', as_index=False).agg({'KitchenSquare': 'mean', 'Square': 'mean'})
    f['%kit'] = f['KitchenSquare'] / f['Square']
    f.drop(['KitchenSquare', 'Square'], axis=1, inplace=True)
    r = r.merge(f, on='HouseYear', how='left')

    r['mhf'] = np.floor(r['mhf'])    
    r['%lif'] = r['%lif'].fillna(r['%lif'].mean())
    r['%kit'] = r['%kit'].fillna(r['%kit'].mean())
    return r


def getSQPrice(data):
    d = data.copy()
    d['SQPrice'] = d['Price'] // d['Square'] // 100 * 100
    return d[d['SQPrice'] < 8000].groupby(['DistrictId', 'Rooms'], as_index=False).agg({'SQPrice': 'mean'})
````

**Добавление средней цены за м^2 по районам / комнатам**

````python
def addSQPrice(data, sqprice):
    median = sqprice['SQPrice'].median()
    data = data.merge(sqprice, on=['DistrictId','Rooms'], how='left')
    data['SQPrice'].fillna(median, inplace=True)
    return data
````

**Иcправление Square, HouseFloor, LifeSquare, KitchenSquare**

````python
def fixFloorAndSquare(data, means):
    data = data.merge(means, on='HouseYear', how='left')
    
    i = data['HouseFloor'] < data['Floor']
    data.loc[i, 'HouseFloor'] = data.loc[i, 'mhf']
    data['Floor'] = data['Floor'].astype(float)
    data['HouseFloor'].fillna(data['Floor'], inplace=True)
    
    ind = fltLifeSquare(data)
    data.loc[ind, 'LifeSquare'] = data.loc[ind, 'Square'] * data.loc[ind, '%lif']
    data['LifeSquare'].fillna(data['LifeSquare'].mean(), inplace=True)

    ind = fltKitchenSquare(data)
    data.loc[ind, 'KitchenSquare'] = data.loc[ind, 'Square'] * data.loc[ind, '%kit']
    data['KitchenSquare'].fillna(data['KitchenSquare'].mean(), inplace=True)

    data.drop(['%kit', '%lif', 'mhf'], axis=1, inplace=True)
    return data
````

**Подготовка данных**

````python
means = getMeans(data)
sqprice = getSQPrice(data)

def prepare(data):
    data = addSQPrice(data, sqprice)
    data = fixFloorAndSquare(data, means)
    data = pd.get_dummies(data)
    return data
    
source = prepare(data)
target = data.loc[:, 'Price']
source.drop('Price', axis=1, inplace=True)

splited = train_test_split(source, target, test_size=0.3, random_state=42)
````

# Выбор модели

````python
# basic linear regression
def runLR(res, feats, scaler=None):
    train, valid, train_price, valid_price = splited
    train = train.loc[:, feats]
    valid = valid.loc[:, feats]

    if scaler:
        train = scaler.fit_transform(train)
        valid = scaler.transform(valid)

    t = time()
    model = LinearRegression(normalize=True)
    model.fit(train, train_price)
    price = model.predict(valid)
    t = round(time()-t, 5)
    res.append(['LinearRegression', r2(valid_price, price), t, ''])


# search by Grid
def runCV(res, feats, scaler=None):
    train = source.loc[:, feats].copy()
    train_price = target.copy()
    if scaler:
        train = scaler.fit_transform(train)
        
    models = {
        'Lasso': ( Lasso(normalize=True, random_state=42), { 'alpha': np.linspace(0.1, 1, 10) } ),
        'Ridge': ( Lasso(normalize=True, random_state=42), { 'alpha': np.linspace(0.1, 1, 10) } ),
        
        'KNeighborsRegressor': (KNeighborsRegressor(weights='distance'), {
            'n_neighbors': np.arange(8,15),
        }),
        
        'DecisionTreeRegressor': (DecisionTreeRegressor(max_features=20, min_samples_leaf=20, random_state=42), {
            'max_depth': np.arange(7, 12),
        }),

        'RandomForest': (RandomForestRegressor(n_estimators=1100, max_features=10, random_state=42), {
            'min_samples_leaf': [2, 3],
            'max_depth': [14, 15, 16],
        })
    }

    for name, item in models.items():
        model, parameters = item
        print(f'Grid search for "{name}":')
        clf = GridSearchCV(n_jobs=-1, estimator=model, param_grid=parameters, scoring='r2', cv=3)
        clf.fit(train, train_price)
        print('\t best:', clf.best_params_)

        results = clf.cv_results_
        for i, t in enumerate(results['mean_fit_time']):
            r = results['mean_test_score'][i]
            p = str(results['params'][i])
            res.append([name, round(r, 7), round(t,7), p])
````

````python
feats = ['Rooms', 'Square', 'LifeSquare', 'KitchenSquare',
       'Floor', 'HouseFloor', 'HouseYear', 'Ecology_1', 'Social_1', 'Social_2',
       'Social_3', 'Helthcare_2', 'Shops_1', 'SQPrice', 'Ecology_2_A',
       'Ecology_2_B', 'Ecology_3_A', 'Ecology_3_B', 'Shops_2_A', 'Shops_2_B']

sc = StandardScaler(with_mean=False)
mx = MinMaxScaler()
rb = RobustScaler()

res = []
runLR(res, feats, mx)
runCV(res, feats, mx)
````

````python
res = pd.DataFrame(res, columns=['model', 'R2', 'Time', 'Test'])
res['Total'] = res['R2'] * 1/res['Time']
res = res.sort_values(by=['R2', 'Total'], ascending=False)
res
````

# Тестирование выбранной модели

* Лучшая модель: **RandomForestRegressor**
* Параметры: ```max_depth=14, min_samples_leaf=2```

````python
scaler = StandardScaler(with_mean=False)
````

````python
train, valid, train_price, valid_price = splited

train = scaler.fit_transform(train.loc[:, feats])
valid = scaler.transform(valid.loc[:, feats])

model = RandomForestRegressor(n_estimators=1100, max_depth=14, min_samples_leaf=2, \
                              max_features=10, random_state=42);
model.fit(train, train_price)
price = model.predict(valid)

r2_score = r2(valid_price, price)
mse_score = mse(valid_price, price)

print(f'R2   = {r2_score:2.5f}\nMSE = {mse_score:2.5f}')
````

### Обучим на полной выборке и предскажем цены для тестовой модели

````python
train = source.loc[:, feats].copy()
train_price = target.copy()
train = scaler.fit_transform(train)
result = prepare(test)

model = RandomForestRegressor(n_estimators=1100, max_depth=14, min_samples_leaf=2, \
                              max_features=10, random_state=42);
model.fit(train, train_price)

result['Price'] = model.predict(scaler.transform(result.loc[:, feats]))
result.loc[:,['Id','Price']].to_csv('SAhranovich_predictions.csv', index=None)
result.head(10)
````
