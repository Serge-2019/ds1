# Подготовка данных

```python
def getMediansByYear(df):
    t = df.loc[df['Rooms'].between(1,9) & df['Square'].between(16,300), \
                  ['Rooms', 'Square', 'LifeSquare', 'KitchenSquare', 'HouseYear', 'HouseFloor', 'Floor']]

    t.loc[(t['KitchenSquare'] > t['Square']*0.3) | (t['KitchenSquare'] <=1), 'KitchenSquare'] = None
    t.loc[t['HouseFloor'] < 1, 'HouseFloor'] = None
    t.loc[t['HouseFloor'] < t['Floor'], 'HouseFloor'] = t.loc[t['HouseFloor'] < t['Floor'], 'Floor']
    
    t = t.groupby('HouseYear').agg({'Rooms': 'median', 'Square': 'median', \
                                    'LifeSquare': 'median', 'KitchenSquare': 'median', 'HouseFloor': 'median'})

    t['Rooms'] = t['Rooms'].astype(np.int64)
    t['SQ_room'] = t['LifeSquare'] / t['Rooms']
    t['%kit'] = t['KitchenSquare'] / t['Square']
    t['%lif'] = t['LifeSquare'] / t['Square']
    t['%kit'] = t['%kit'].fillna(t['%kit'].mean())
    t['%lif'] = t['%lif'].fillna(t['%lif'].mean())
    t['KitchenSquare'] = t['KitchenSquare'].fillna(t['KitchenSquare'].median())

    return t

md = getMediansByYear(data)
```

```python
def fixData(df, md):
    t = df.copy()
    
    # обновим все выбросы по площади - поставим их равным медиане
    a = df.loc[(df['Square'] > 300) | (df['Square'] < 15), \
               ['Id', 'HouseYear', 'Square', 'Rooms', 'LifeSquare', 'KitchenSquare']]
    a = a.set_index('HouseYear')
    a.update(md)
    a = a.set_index('Id')
    t = t.set_index('Id')
    t.update(a)
    t = t.reset_index()

    # обновим все выбросы по HouseFloor - поставим их равным медиане
    a = df.loc[(df['HouseFloor'].isnull()) | (df['HouseFloor'] < 1), \
               ['Id', 'HouseYear', 'HouseFloor']]
    a = a.set_index('HouseYear')
    a.update(md)
    a = a.set_index('Id')
    t = t.set_index('Id')
    t.update(a)

    s = md.drop(['Rooms', 'Square', 'LifeSquare', 'KitchenSquare', 'HouseFloor'], axis=1)
    t = t.reset_index().merge(s, how='left', on='HouseYear')
    t['%kit'] = t['%kit'].fillna(s['%kit'].mean())
    t['%lif'] = t['%lif'].fillna(s['%lif'].mean())
    t['SQ_room'] = t['SQ_room'].fillna(s['SQ_room'].mean())
    
    # исправим KitchenSquare и LifeSquare сделаем их более реальными
    ind = (t['KitchenSquare'] > t['Square']*0.3) | (t['KitchenSquare'] <=1)
    t.loc[ind, 'KitchenSquare'] = t.loc[ind, '%kit'] * t.loc[ind, 'Square']
    
    ind = (t['Square'] < t['LifeSquare']) | (t['LifeSquare'].isnull()) | (t['LifeSquare']< t['Square']*0.2)
    t.loc[ind, 'LifeSquare'] = t.loc[ind, '%lif'] * t.loc[ind, 'Square']

    # исправим выбросы по Rooms
    ind = (t['Rooms'] < 1) | (t['Rooms'] > 8)
    t.loc[ind, 'Rooms'] = t.loc[ind, 'Square'] // t.loc[ind, 'SQ_room']

    # исправим выбросы по Rooms
    ind = t['HouseFloor'] < t['Floor']
    t.loc[ind, 'HouseFloor'] = t.loc[ind, 'Floor']

    t.drop(['%kit', '%lif', 'SQ_room'], axis=1, inplace=True)
    return t

# data = fixData(data, md)
# test = fixData(test, md)
```

# Тестирование модели

```python
feats = ['Rooms', 'Square', 'LifeSquare', 'KitchenSquare', \
         'Floor', 'HouseFloor', \
         'DistrictId', \
         'HouseYear']

X = data.loc[:, feats]
y = data.loc[:, 'Price']

ind = X['LifeSquare'].isnull()
X.loc[ind,'LifeSquare'] = X['LifeSquare'].mean()

splitdata = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
def run(res, splitdata, pre=''):
    train, valid, train_price, valid_price = splitdata

    models = {
        'LinearRegression': LinearRegression(),
        'Lasso a = 20': Lasso(alpha=20, normalize=True),
        'Ridge a = 19000': Ridge(alpha=190000),
        'Ridge a = 0.2': Ridge(alpha = 0.2),
        'RandomForest': RandomForestRegressor(max_depth=20, min_samples_leaf=3, \
                                              random_state=42, max_features='log2', n_estimators=1000)
    }

    for name, model in models.items():
        model.fit(train, train_price)
        price = model.predict(valid)
        res.append([pre + name, r2(valid_price, price), mse(valid_price, price)])

    return models
```

```python
res = []

models = run(res, splitdata)
```

```python
from sklearn.preprocessing import MinMaxScaler

train, valid, train_price, valid_price = splitdata

scaler = MinMaxScaler()
train = pd.DataFrame(scaler.fit_transform(train, train_price), columns=train.columns)
valid = pd.DataFrame(scaler.transform(valid), columns=valid.columns)

ns = (train, valid, train_price, valid_price)
```

```python
models = run(res, ns, 'Scaled: ')
```
